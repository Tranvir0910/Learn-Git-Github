Một hệ thống tích hợp liên tục sẽ xây dựng và 
kiểm tra mã của chúng tôi mỗi khi có thay đổi.

Sau khi mã của chúng tôi được 
tạo và kiểm tra tự động, bước tự động hóa tiếp theo 
là triển khai liên tục, đôi khi được gọi là phân phối liên tục hoặc CD.

Triển khai liên tục có nghĩa là mã 
mới được triển khai thường xuyên. Mục tiêu là để tránh triển khai với nhiều thay đổi giữa hai phiên bản của một dự án và 
thay vào đó thực hiện các cập nhật gia tăng chỉ với một vài thay đổi tại một thời điểm.

Điều này cho phép phát hiện và sửa lỗi sớm.


Có rất nhiều khái niệm mà bạn sẽ cần 
xử lý khi tạo CI/CD của riêng mình.

Đầu tiên là một khái niệm về **Pipelines**.

Pipelines chỉ định các bước cần 
chạy để nhận được kết quả bạn muốn.

Đối với một Dự án Python đơn giản, quy trình có thể chỉ chạy thử nghiệm tự động.

Đối với một dịch vụ web được viết bằng Go, quy trình có thể được biên dịch chương trình, chạy thử nghiệm đơn vị và thử nghiệm tích hợp và cuối 
cùng là triển khai mã cho một phiên bản thử nghiệm.

Một khái niệm khác xuất hiện 
khi thực hiện CI/CD là **Artifacts**.

Đây là tên được sử dụng để mô tả bất kỳ tệp 
nào được tạo như một phần của quy trình.

Điều này thường bao gồm các 
phiên bản được biên dịch của mã nhưng có thể bao gồm các tệp được tạo khác như tệp PDF cho 
tài liệu hoặc gói dành riêng cho hệ điều hành để cài đặt dễ dàng.

Khi thiết lập CI/CD, chúng ta phải cẩn 
thận về cách chúng tôi quản lý các bí mật.

Nếu quy trình của chúng tôi bao gồm việc triển khai 
một phiên bản phần mềm mới cho máy chủ thử nghiệm, thì bằng cách nào đó, chúng tôi cần cấp cho phần mềm đang chạy 
quy trình quyền truy cập vào máy chủ thử nghiệm của mình.

Có nhiều chiến lược khác 
nhau để thực hiện việc này, chẳng hạn như **trao đổi khóa SSH** hoặc **sử dụng mã thông báo API** dành riêng cho ứng dụng.

Hai điều cần nhớ, trước tiên:

Hãy đảm bảo rằng các thực thể được ủy quyền cho máy chủ thử nghiệm không 
giống với các thực thể được ủy quyền triển khai trên máy chủ sản xuất.

Bằng cách đó, nếu có bất kỳ hình thức 
thỏa hiệp nào trong quy trình, thì máy chủ sản xuất của 
bạn sẽ không bị ảnh hưởng.

Thứ hai, luôn có kế hoạch khôi phục quyền truy cập của 
bạn trong trường hợp đường dẫn của bạn bị xâm phạm.
