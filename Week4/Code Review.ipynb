### **What are code reviews?**

Thực hiện đánh giá mã có nghĩa là xem qua mã, tài liệu hoặc cấu hình của người khác và kiểm tra xem 
tất cả có hợp lý và tuân theo các mẫu dự kiến hay không.

Mục tiêu của việc xem xét mã là cải thiện dự án bằng 
cách đảm bảo rằng các thay đổi có chất lượng cao.
