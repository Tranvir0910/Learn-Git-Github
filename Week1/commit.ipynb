## **Git commit**

Một commit thường được chia thành vài phần:

- phần đầu là một bản tóm tắt ngắn về commit theo sau là một dòng trống.

- Tiếp theo là phần mô tả đầy đủ về các thay đổi nêu chi tiết lý do tại sao chúng lại cần thiết và bất kỳ điều gì có thể đặc biệt thú vị hoặc khó hiểu về chúng.

- Sau dòng đầu tiên là một dòng trống và phần còn lại của văn bản thường được giữ dưới 72 ký tự.


Có một lệnh git được sử dụng để hiển thị các thông báo cam kết này được gọi là ***git log.***
