## **Using Git**


Sử dụng ***git config*** để cho biết chúng ta là ai

```python
C:\Users\TranVi> git config --global user.email "example@gmail.com"

C:\Users\TranVi> git config --global user.name "example"
```

thêm vào file **.gitconfig** tại C:\Users\Example\

- Tạo thư mục checks

- Tạo kho lưu trữ đầu tiên với ***git init***

- ***git clone*** để copy 1 repository

- ***ls -la*** để check xem thư mục này có tồn tại hay không hay liệt kê các tệp bắt đầu bằng dấu chấm

- ***ls -l .git/*** để kiểm tra bên trong thư mục .git có gì

- ***Thư mục .git*** You can think of it as a database for your Git project that stores the changes and the change history.

#### **Working tree**

- Thư mục bên ngoài .git là cây làm việc

- Cây làm việc là phiên bản hiện tại của dự án

- Cây làm việc chứa tất cả các tệp hiện đang được git theo dõi 

- **cp ../disk_usage.py .** copy file thêm file vào working tree **nhưng không được theo dõi**

- ***git add disk_usage.py*** thêm file vào dự án và được git theo dõi, cùng với đó tệp đã được thêm vào ***Staging Area***

- **git status** kiểm tra trạng thái

Khu vực tổ chức còn được gọi là chỉ mục là một tệp do Git duy trì chứa tất cả thông tin về những tệp và thay đổi nào sẽ đi vào lệnh tiếp theo của bạn.

Để đưa vào Repository chúng ta sử dụng ***git commit***

***Working tree ( files ) --> Staging Area ( Vận chuyển file ) --> Git directory (.git)***
