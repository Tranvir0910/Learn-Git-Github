### **Merging**

Hp nht l thut ng m Git s dng  kt 
hp d liu nhnh v lch s vi nhau.

Chng ta s s dng lnh git merge, lnh ny cho php chng ta 
ly cc nh chp nhanh v lch s c lp ca mt nhnh Git, ri trn chng vo mt nhnh khc.

Git s dng hai thut ton khc nhau  thc hin hp nht, **chuyn tip nhanh v hp nht ba chiu**. 

Hp nht m chng ti va thc hin l mt v d v hp nht chuyn tip nhanh. 

- Kiu hp nht ny xy ra khi tt c cc xc nhn trong nhnh  kim tra cng nm trong nhnh ang c hp nht. Nu y l trng hp, chng ta c th ni rng lch s commit ca c hai nhnh khng khc nhau. 

- Trong nhng trng hp ny, tt c nhng g Git phi lm l cp nht cc con tr ca cc nhnh thnh cng mt cam kt v khng cn phi hp nht thc s. 

Mt khc, hp nht ba chiu c thc hin khi lch s ca cc nhnh hp nht  phn k theo mt cch no  v khng c ng dn tuyn tnh p  kt hp chng thng qua chuyn tip nhanh. iu ny xy ra khi mt cam kt c thc hin trn mt nhnh sau thi im c hai nhnh tch ra.
