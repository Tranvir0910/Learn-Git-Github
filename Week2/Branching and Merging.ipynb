### **What is a branch?**

Trong Git, một nhánh ở mức cơ bản nhất 
chỉ là một con trỏ tới một cam kết cụ thể.

Nhánh mặc định mà Git tạo cho bạn khi khởi 
tạo kho lưu trữ mới được gọi là master.

Master thường được sử dụng để biểu 
diễn trạng thái tốt đã biết của một dự án.


### **Creating New Branches**

Các nhánh rất cần thiết cho cách thức hoạt động được thực hiện trong GIT, có rất nhiều cách khác nhau để làm việc với chúng.

#### **git branch**

Câu lệnh cho chúng ta biết chúng ta đang ở nhánh nào bằng cách hiển thị dấu '*' trước tên nhánh và có bao nhiêu nhánh

#### **Tạo một nhánh mới: git branch < name_branch >**

Tạo một nhánh mới

#### **Di chuyển đến nhánh mới: git check out < name_branch >**

#### **Tạo một branch và di chuyển đến branch: git checkout -b < name_branch >**

#### **Thêm file vào nhánh mới và commit**

Như ta đã thấy HEAD đã chuyển đến nhánh mới được tạo ( HEAD --> even-better-feature )

### **Working with Branch**

Sau khi chúng ta chuyển nhánh, commit bên nhánh **even-better-feature** đã không còn.

Nhờ kiểm tra này, head đã chuyển từ việc trỏ đến cam kết mới nhất trong nhánh tính năng thậm chí còn tốt hơn sang cam kết gần đây nhất của nhánh chính.

#### **Xóa một Branch: git branch -d < name_branch >**

Nếu có những thay đổi trong nhánh mà chúng ta muốn 
xóa mà chưa được hợp nhất lại vào nhánh chính, thì nó sẽ báo lỗi

### **Merging**

Hợp nhất là thuật ngữ mà Git sử dụng để kết 
hợp dữ liệu nhánh và lịch sử với nhau.

Chúng ta sẽ sử dụng lệnh git merge, lệnh này cho phép chúng ta 
lấy các ảnh chụp nhanh và lịch sử độc lập của một nhánh Git, rồi trộn chúng vào một nhánh khác.

Git sử dụng hai thuật toán khác nhau để thực hiện hợp nhất, **chuyển tiếp nhanh và hợp nhất ba chiều**. 

Hợp nhất mà chúng tôi vừa thực hiện là một ví dụ về hợp nhất chuyển tiếp nhanh. 

- Kiểu hợp nhất này xảy ra khi tất cả các xác nhận trong nhánh đã kiểm tra cũng nằm trong nhánh đang được hợp nhất. Nếu đây là trường hợp, chúng ta có thể nói rằng lịch sử commit của cả hai nhánh không khác nhau. 

- Trong những trường hợp này, tất cả những gì Git phải làm là cập nhật các con trỏ của các nhánh thành cùng một cam kết và không cần phải hợp nhất thực sự. 

Mặt khác, hợp nhất ba chiều được thực hiện khi lịch sử của các nhánh hợp nhất đã phân kỳ theo một cách nào đó và không có đường dẫn tuyến tính đẹp để kết hợp chúng thông qua chuyển tiếp nhanh. Điều này xảy ra khi một cam kết được thực hiện trên một nhánh sau thời điểm cả hai nhánh tách ra.
