### **git commit -a**

```python
git commit -a
```

git commit -a là lối tắt để sắp xếp bất kỳ thay đổi nào đối với các tệp được theo dõi và cam kết chúng trong một bước.

sử dụng git commit -a sẽ bỏ qua staging area

Có nghĩa là bạn không thể thêm bất kỳ 
thay đổi nào khác trước khi tạo cam kết. Vì vậy, bạn cần chắc chắn rằng bạn đã bao 
gồm mọi thứ bạn muốn đưa vào cam kết đó.

### **HEAD --> Master**

***HEAD --> Master***

Git sử dụng bí danh head để thể hiện ảnh chụp nhanh (**current snapshot**) dự án của bạn hiện đã được kiểm tra.

Chúng ta đã thấy cách git log hiển thị cho chúng ta danh sách 
các cam kết được thực hiện trong kho lưu trữ Git hiện tại. Theo mặc định, nó in thông báo cam 
kết, tác giả và ngày thay đổi. có lẽ chúng ta cũng cần xem xét các dòng 
thực tế đã thay đổi trong mỗi lần xác nhận.

### **git log -p**

```python
git log -p ( patch )
```

p xuất phát từ patch, bởi vì sử dụng cờ này sẽ cho chúng ta patch đã được tạo.

### **git show + ID**

Nếu chúng ta không muốn cuộn xuống cho đến khi tìm thấy cam kết mà chúng tôi thực sự quan tâm, một tùy chọn khác là sử dụng lệnh ***git show***.

Lệnh này lấy ID cam kết làm tham số và sẽ hiển 
thị thông tin về cam kết và bản vá liên quan. ID là mã định danh mà chúng 
ta thấy bên cạnh từ cam kết trong nhật ký.

### **git log --stat**

- ***git log --stat*** Điều này sẽ khiến nhật ký git hiển thị một 
số thống kê về các thay đổi trong cam kết, chẳng hạn như tệp nào đã được thay đổi 
và số lượng dòng đã được thêm hoặc xóa.

### **git diff**

Trước khi cam kết, bạn cần đảm bảo rằng nó hoạt động chính xác. Kiểm tra xem tất cả các trường hợp thử nghiệm đã được bao phủ chưa, v.v. Vì vậy, trong khi làm điều này, bạn sẽ tìm thấy các lỗi trong mã của mình mà bạn cần sửa. 

Điều tự nhiên là khi bạn đến bước cam kết, bạn không thực sự nhớ mọi thứ bạn đã thay đổi. Để giúp chúng tôi theo dõi, git cung cấp cho chúng tôi lệnh ***git diff***.

### **git add -p**

Một việc khác mà chúng ta có thể làm để xem lại các thay đổi 
trước khi thêm chúng là sử dụng cờ ***-p với lệnh git add***.
