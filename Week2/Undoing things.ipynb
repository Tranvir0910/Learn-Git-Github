### **Rollbacks**

**"git revert"** l mt lnh c s dng  hon tc cc thay i c thc hin trong mt commit c th. 

Khi bn thc hin mt "git revert," Git s to mt commit mi cha cc thay i ngc li ca commit bn mun hon tc.

Khi bn hon tc mt commit bng "git revert," cc thay i trong commit  s khng b xa khi lch s ca repository. 

Thay vo , Git s thm mt commit mi  "hon tc" cc thay i  thay i trong commit gc.

git  to 1 commit ngc li vi commit trc .
