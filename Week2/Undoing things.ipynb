#### **git add * .**

chúng ta tạo một tệp tạm thời với 
đầu ra của tập lệnh của chúng tôi.

Thêm vào working tree bằng lệnh **( git add * )**

Tạo đầu ra vào file **test.txt**.

Sau đó chúng ta sử dụng **( git add * )** để thêm những tệp có trong thư mục ( working tree ) nhưng không thêm những tệp hoặc thư mục bắt đầu bằng dấu .

**Lưu Ý**:
- **( git add * )** có nghĩa là thêm tất cả các tệp trong thư mục hiện tại, ngoại trừ các tệp có tên bắt đầu bằng dấu chấm. 
- **( git add . )** thêm toàn bộ thư mụcbao gồm cả các tệp có tên bắt đầu bằng dấu chấm, 2 lệnh gần như gần như giống nhau.

### **Amending Commits**

#### **git commit --amend**

Git sẽ lấy bất cứ thứ gì hiện có trong Staging Area của chúng ta và

nhưng bạn không nên sử dụng nó cho các cam kết công khai.

### **Rollbacks**

**"git revert"** là một lệnh được sử dụng để hoàn tác các thay đổi được thực hiện trong một commit cụ thể. 

Khi bạn thực hiện một "git revert," Git sẽ tạo một commit mới chứa các thay đổi ngược lại của commit bạn muốn hoàn tác.

Khi bạn hoàn tác một commit bằng "git revert," các thay đổi trong commit đó sẽ không bị xóa khỏi lịch sử của repository. 

Thay vào đó, Git sẽ thêm một commit mới để "hoàn tác" các thay đổi đã thay đổi trong commit gốc.

git đã tạo 1 commit ngược lại với commit trước đó.
